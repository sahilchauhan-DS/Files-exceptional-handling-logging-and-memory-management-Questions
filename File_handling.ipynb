Que.1 What is the difference between interpreted and compiled languages?

  ->  The key difference between interpreted and compiled languages lies in how they are processed: compiled languages are translated into machine code before execution, while interpreted languages are translated and executed line by line during runtime.

  Compiled Languages:

Translation:
The entire program is translated into machine code (executable code) by a compiler before it is run.

Execution:
The compiled machine code is then executed directly by the computer's processor.

Speed:
Compiled languages generally offer faster execution speeds because the translation process happens only once, before running the program.

Examples:
C, C++, and Java (when compiled).

Error Detection:

Interpreted Languages:

Translation: The code is translated and executed line by line by an interpreter during runtime.

Execution: The interpreter reads and executes each line of code, one at a time.

Speed: Interpreted languages can be slower than compiled languages because the translation process happens each time the program is run.

Examples: Python, JavaScript, and PHP.

Error Detection: Errors are usually detected and reported during runtime, which can make debugging easier.

Flexibility: Interpreted languages offer more flexibility for testing and modifying code on the fly.

Que.15 How does Python's garbage collection system work?

   ->  Python's garbage collection system uses a hybrid approach of reference counting and generational garbage collection to automatically manage memory, reclaiming memory occupied by objects that are no longer in use, which prevents memory leaks.

   1. Reference Counting:
Each object in Python has a reference count, which tracks how many other objects are referring to it.
When an object is created, its reference count is set to 1.
When another variable or data structure refers to the same object, the reference count is incremented.
When a reference to an object is removed, the reference count is decremented.
When an object's reference count drops to 0, the memory occupied by that object is immediately reclaimed.

2. Generational Garbage Collection:
Python uses a generational garbage collector to handle cyclic references (where objects reference each other, preventing reference counting from working correctly).
The garbage collector classifies objects into three generations (0, 1, and 2) based on how many collection sweeps they have survived.
New objects start in generation 0, and if they survive a collection, they move to a higher generation.
The garbage collector periodically scans objects in the youngest generation (generation 0) and checks for unreachable objects.
If an object is part of a circular reference but cannot be accessed by any active references, it is collected.
Objects that survive this process are promoted to the next generation, where they are scanned less frequently.
The garbage collector keeps track of the number of object allocations and deallocations since the last collection.
When the number of allocations minus the number of deallocations exceeds a threshold, collection starts.

3. The gc Module:
Python provides the gc module, which provides tools to inspect and manipulate the garbage collection process.
You can use functions in this module to enable or disable garbage collection, set thresholds, and inspect collected objects.

```python
#Que.4 Write a Python script that reads from one file and writes its content to another file?

def copy_file(source_file, destination_file):
    try:
        with open(source_file, 'r') as src:
            content = src.read()

        with open(destination_file, 'w') as dest:
            dest.write(content)

        print(f"Contents copied from '{source_file}' to '{destination_file}' successfully.")

    except FileNotFoundError:
        print(f"Error: The file '{source_file}' does not exist.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

# Example usage:
source = "input.txt"
destination = "output.txt"
copy_file(source, destination)
```

```python
#Que.18 How would you implement a basic logging setup that logs to a file with rotation after 1MB?

import logging
from logging.handlers import RotatingFileHandler

# Configure the logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.DEBUG)

# Set up a rotating file handler
handler = RotatingFileHandler(
    "app.log",         # Log file name
    maxBytes=1_000_000,  # Rotate after 1MB
    backupCount=3        # Keep 3 backup log files
)

# Log message format
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example log messages
for i in range(10000):
    logger.info(f"Logging line {i}")
```

```python
#Que.21 Write a Python program that reads a file and prints the number of occurrences of a specific word?

def count_word_occurrences(filename, target_word):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            # Convert to lowercase for case-insensitive search
            words = content.lower().split()
            count = words.count(target_word.lower())
            print(f"The word '{target_word}' occurs {count} time(s) in the file.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

# Example usage
count_word_occurrences("sample.txt", "python")
```

```python
import logging

# Configure logging
logging.basicConfig(
    filename='error_log.txt',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            contents = file.read()
            print(contents)
    except FileNotFoundError as e:
        logging.error(f"File not found: {file_path} - {e}")
        print("Error: File not found.")
    except IOError as e:
        logging.error(f"IO error when accessing file: {file_path} - {e}")
        print("Error: IO error occurred.")
    except Exception as e:
        logging.error(f"Unexpected error: {e}")
        print("An unexpected error occurred.")

# Example usage
if __name__ == "__main__":
    file_path = input("Enter the file path to read: ")
    read_file(file_path)
```
